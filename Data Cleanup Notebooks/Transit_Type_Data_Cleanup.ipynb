```python
# import Dependencies
import os 
import pandas as pd 
import numpy as np
```

```python
# Create a reference to the CSV and import it into a Pandas DataFrame

csv_path = "../ELT_CSVs/typetransit2011.csv"
raw_transit_df = pd.read_csv(csv_path)
raw_transit_df.head()
```

```python
# Begin dropping unnecessary columns
margin_transit_df = raw_transit_df[raw_transit_df.columns.drop(list(raw_transit_df.filter(regex='Margin')))]
percent_impunted_transit_df = margin_transit_df[margin_transit_df.columns.drop(list(margin_transit_df.filter(regex='PERCENT IMPUTED')))]
race_transit_df = percent_impunted_transit_df[percent_impunted_transit_df.columns.drop(list(percent_impunted_transit_df.filter(regex='HISPANIC')))]
vehicles_available_transit_df = race_transit_df[race_transit_df.columns.drop(list(race_transit_df.filter(regex='VEHICLES AVAILABLE')))]
housing_tenure_df = vehicles_available_transit_df[vehicles_available_transit_df.columns.drop(list(vehicles_available_transit_df.filter(regex='HOUSING TENURE')))]
gender_df = housing_tenure_df[housing_tenure_df.columns.drop(list(housing_tenure_df.filter(regex='SEX')))]
travel_time_df = gender_df[gender_df.columns.drop(list(gender_df.filter(regex='TRAVEL TIME')))]
median_df = travel_time_df[travel_time_df.columns.drop(list(travel_time_df.filter(regex='Median')))]
industry_df = median_df[median_df.columns.drop(list(median_df.filter(regex='INDUSTRY')))]
occupation_df = industry_df[industry_df.columns.drop(list(industry_df.filter(regex='OCCUPATION')))]
language_df = occupation_df[occupation_df.columns.drop(list(occupation_df.filter(regex='LANGUAGE')))]
nativity_df = language_df[language_df.columns.drop(list(language_df.filter(regex='NATIVITY')))]
race_df = nativity_df[nativity_df.columns.drop(list(nativity_df.filter(regex='race')))]
time_leaving_df = race_df[race_df.columns.drop(list(race_df.filter(regex='TIME LEAVING')))]
place_work_df = time_leaving_df[time_leaving_df.columns.drop(list(time_leaving_df.filter(regex='PLACE')))]
class_df = place_work_df[place_work_df.columns.drop(list(place_work_df.filter(regex='CLASS')))]
at_home_df = class_df[class_df.columns.drop(list(class_df.filter(regex='households')))]
poverty_df = at_home_df[at_home_df.columns.drop(list(at_home_df.filter(regex='POVERTY')))]
earnings_df = poverty_df[poverty_df.columns.drop(list(poverty_df.filter(regex='EARNINGS')))]
white_df = earnings_df[earnings_df.columns.drop(list(earnings_df.filter(regex='White')))]
over_df = white_df[white_df.columns.drop(list(white_df.filter(regex='over.1')))]
age_df = over_df[over_df.columns.drop(list(over_df.filter(regex='AGE')))]

# Show the Header 
age_df.head()
```

```python
# Drop More Columns
full_geography_df = age_df[age_df.columns.drop('Id')]
only_ID2_df = full_geography_df[full_geography_df.columns.drop('Geography')]

# Rename ID2 Column to Zipcode
only_zipcode_df = only_ID2_df.rename(index=str, columns={"Id2": "Zipcode"})
only_zipcode_df.head()
```

```python
# Create new csv file to examine
only_zipcode_df.to_csv('Transit_Type_Semi_Cleaned.csv', encoding='utf-8', index=False)
```
